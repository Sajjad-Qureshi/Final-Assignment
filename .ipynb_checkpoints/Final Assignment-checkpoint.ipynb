#### Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Reading Dataset 

```python
heart_df = pd.read_csv('../Dataset/heart.csv')
```

```python
# Displaying Few Rows from Dataset
heart_df.head()
```

#### Some Stats

```python
heart_df.info()
```

```python
heart_df.describe()
```

```python
# Checking for any null values
heart_df.isnull().sum()
```

```python
# Replacing 0,1 in target with negative and Positive for better understanding
heart_df.target.replace([0,1],["Negative","Positive"],inplace=True)

# Replacing 0,1 in sex with F and M for better understanding
heart_df.sex.replace([0,1],["F","M"],inplace=True)
```

#### Exploratory Data Analysis

#### 1. Positive and Negative Patients For Heart Diseases Graphes

```python
# Groupingby target Column and taking age column for values and saving it in result, You can also pick anyother column instead of age column!! 
result = heart_df.groupby('target').count()['age']

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of Positive and Negative Patients with Heart Disease!',size=15)

sns.barplot(ax=axes[0], x=result.index, y=result.values)
plt.pie(result.values,labels=result.index,autopct='%.0f%%');
```

#### 2. Number of Females and Males in Dataset!

```python
# Groupingby sex Column and taking age column for values and saving it in result, You can also pick anyother column instead of age column!
result = heart_df.groupby('sex').count()['age']

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of Females and Males in Dataset!',size=15)

sns.barplot(ax=axes[0], x=['Female','Male'], y=result.values)
plt.pie(result.values,labels=['Female','Male'],autopct='%.0f%%');
```

#### 3. Number of positive Male and Female Patients for Heart Disease!

```python
# Groupingby [sex and target] Column and taking age column for values and saving it in result, 
# 'Iloc' funtion is used to filter only positive patients number, 
# You can also pick anyother column instead of age column!

result = heart_df.groupby(['sex','target']).count()['age'].iloc[1:4:2]

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of Positive Male and Female patients for Heart Disease!',size=15)

sns.barplot(ax=axes[0], x=['Female Positive','Male Positive'], y=result.values)
plt.pie(result.values,labels=['Female Positive','Male Positive'],autopct='%.0f%%');
```

#### 4. Which age group has most Positive Number?

```python
# Filtering only Positive Cases from Dataset!
result = heart_df[heart_df['target'] == 'Positive']

# Particular Age
age_group = result.groupby('age').count()['sex'].index

# Number of Positive cases Particular Age Wises
positive_cases_number = result.groupby('age').count()['sex'].values

# Adding positive Cases numbers in 10s Group
x_axis = ['10s','20s','30s','40s','50s','60s','70s','80s','90s','100s']
total_pos_num = []
for i in range(10,101,10):
    addition = 0
    for j in range(0,40):
        if age_group[j] > i-10 and age_group[j] <= i:
            addition += positive_cases_number[j]     
    total_pos_num.append(addition)

# Ploting a Barplot Showing Age Groups and Number of Positive Cases according to that age group!
sns.barplot(x=x_axis, y=total_pos_num)
plt.title(label='Positive Cases Number Age Group Wise',size=15)
plt.xlabel('Age in 10s',size=13)
plt.ylabel('Number of Positive Cases',size=13);
```

#### 5. Number of People Having Different Chest Pain!

```python
# Groupingby Cp (Chest Pain) Column and taking age column for values and saving it in result, You can also pick anyother column instead of age column!
result = heart_df.groupby('cp').count()['age']

# In Our Data Set Chest Pain is reprented in Number values from (0-3), Hence Changing Number Values into Their Real Names!
chest_pain_types = ['Asymptomatic','Atypical angina','Non-anginal pain','Typical angina']

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of People Having Different Chest Pain!',size=15)

ax = sns.barplot(ax=axes[0], x=chest_pain_types, y=result.values)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 60)
plt.pie(result.values,labels=chest_pain_types,autopct='%.0f%%');
```

#### 6. Number of Diabetic and Non-Diabetic people!

```python
# Groupingby FBS (Fasting Blood Sugar) Column and taking age column for values and saving it in result, You can also pick anyother column instead of age column!
result = heart_df.groupby('sex').count()['age']

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of Diabetic and Non-Diabetic people!',size=15)

sns.barplot(ax=axes[0], x=['Diabetic','Not-Diabetic'], y=result.values)
plt.pie(result.values,labels=['Diabetic','Not-Diabetic'],autopct='%.0f%%');
```

#### 7. Number of People having conditions according to their Resting Echocardiographic Result!

```python
# Groupingby restecg (Resting Echocardiographic result) Column and taking age column for values and saving it in result, You can also pick anyother column instead of age column!
result = heart_df.groupby('restecg').count()['age']

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of People having conditions according to their Resting Echocardiographic Result!!',size=15)

ax = sns.barplot(ax=axes[0], x=['left ventricular hypertrophy','normal','ST-T wave abnormality'], y=result.values)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.pie(result.values,labels=['left ventricular hypertrophy','normal','ST-T wave abnormality'],autopct='%.0f%%');
```

#### 8. Number of people having different number of Major Vessels!

```python
# Groupingby FBS (Fasting Blood Sugar) Column and taking age column for values and saving it in result, You can also pick anyother column instead of age column!
result = heart_df.groupby('ca').count()['age']

# Ploting a multi-plot with barplot and a pie plot!
fig, axes = plt.subplots(ncols=2, figsize=(13, 5))
fig.suptitle('Number and Ratio of people having different number of Major Vessels!',size=15)

sns.barplot(ax=axes[0], x=result.index, y=result.values)
plt.pie(result.values,labels=result.index,autopct='%.0f%%');
```
